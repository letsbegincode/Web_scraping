```python
from bs4 import BeautifulSoup

# Read the HTML content from the file
with open("file-output.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")


```

```python
# print(soup)
```

```python
blocks=soup.find_all('div',{'class':'card'})
```

```python
len(blocks)
```

```python
print(blocks[0])
```

```python
Name = []
Author = []
Old_Price = []
New_Price = []
Ratings = []
current_offers = []

for block in blocks:
    # Extract text safely with error handling
    name_elem = block.find('span', class_='booktitle')
    author_elem = block.find('span', class_='author')
    old_price_elem = block.find('span', class_='actualprice')
    new_price_elem = block.find('span', class_='initialprice')
    ratings_elem = block.find('span', class_='averageratingbox')
    offer_elem = block.find('div', class_='offer')

    # Use default values if elements are not found
    Name.append(name_elem.get_text(strip=True) if name_elem else 'N/A')
    Author.append(author_elem.get_text(strip=True) if author_elem else 'N/A')
    Old_Price.append(old_price_elem.get_text(strip=True) if old_price_elem else 'N/A')
    New_Price.append(new_price_elem.get_text(strip=True) if new_price_elem else 'N/A')
    Ratings.append(ratings_elem.get_text(strip=True) if ratings_elem else 'N/A')
    current_offers.append(offer_elem.get_text(strip=True) if offer_elem else 'N/A')
```

```python
import pandas as pd

df=pd.DataFrame({'Name':Name,'Author':Author,'Old_Price':Old_Price,'New_Price':New_Price,'Ratings':Ratings,'current_offers':current_offers})
```

```python
df.head()
```

```python
temp_df=df[df['New_Price']=='N/A']
```

```python
temp_df['New_Price']=temp_df['Old_Price']
temp_df['current_offers']=0
```

```python
df['New_Price']=df['New_Price'].str.replace('₹','').str.replace(',','').astype(int)
df['Old_Price']=df['Old_Price'].str.replace('₹','').str.replace(',','').astype(int)
```

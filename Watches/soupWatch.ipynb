```python
import requests
```

```python
r= requests.get('https://www.maximawatches.com/collections/smart-watches')
r.encoding = 'utf-8'
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')

soup.prettify()
```

```python
divs = soup.find_all('div')
len(divs)
```

```python
watches=soup.find_all('div','grid-item')
len(watches)
```

```python
print(watches[1])
```

```python
variants=[]

for watch in watches:
   temp= watch.find_all('div',"tooltip")
   st=""
   for t in temp:
     st=st+t.text.strip()+","
   variants.append(st.replace("\n","").strip())

variants[0]
```

```python
from types import new_class
discounts=[]
old_prices=[]
new_prices=[]
names=[]

for watche in watches:
  # print(watche.prettify())
  discount = watche.find('div','product-label').get_text(strip=True)
  discounts.append(discount)

  name = watche.find('a', class_='product-title').get_text(strip=True)
  old_price = watche.find('span','old-price').get_text(strip=True)
  new_price = watche.find('span','special-price').get_text(strip=True)
  names.append(name)

  old_prices.append(old_price)
  new_prices.append(new_price)
```

```python
import pandas as pd
df=pd.DataFrame()
df['name']=names
df['discount(%)']=discounts
df['old_price']=old_prices
df['new_price']=new_prices
df['variants']=variants

df.head()
```

```python
df['discount(%)']=df['discount(%)'].str.replace('-','')
df['discount(%)']=df['discount(%)'].str.replace('%','').str.replace('Sold Out','').astype(int)

```

```python
df['old_price']=df['old_price'].str.replace('Rs. ','').str.replace(',',"").astype(float).astype(int) # Convert to float first to handle decimals, then to int
df['new_price']=df['new_price'].str.replace('Rs. ','').str.replace(',',"").astype(float).astype(int) # Convert to float first to handle decimals, then to int
```
